```python
import numpy as np
import pickle, os
import matplotlib.pyplot as plt
```

```python
def plot_numpy(array, title='Array plot', save_dir='', rot_num=0):
    plt.figure(figsize=(10, 4))
    array = np.rot90(array, rot_num)
    plt.imshow(array, aspect='auto', origin='lower')
    plt.title(title)
    if save_dir != '':
        plt.savefig(save_dir)
```

```python
file_name = 'm5_scales_breathy_e.npy'
librispeech_file_path = os.path.join(root_path,'librispeech_train_clean_100/LibriSpeech_train-clean-100_2911/7601_2911-7601-0022.npy')

singer_id_dic = {'m1_':'_male1_','m2_':'_male2','m3_':'_male3','m4_':'_male4',
                 'm5_':'_male5','m6_':'_male6','m7_':'_male7','m8_':'_male8',
                 'm9_':'_male9','m10':'_male10','m11':'_male11',
                 'f1_':'female1','f2_':'female2','f3_':'female3','f4_':'female4',
                 'f5_':'female5','f6_':'female6','f7_':'female7','f8_':'female8','f9_':'female9'}

root_path = os.path.join(os.getcwd(), 'enc_preproc_outputs')
parent_dir = 'train/' +singer_id_dic[file_name[:3]]
npss_arr = np.load(os.path.join(root_path, 'vocalset_subset_npss', parent_dir, file_name))
rtvc_arr = np.load(os.path.join(root_path, 'full_vocalset', parent_dir, file_name))
autovc_arr = np.load(os.path.join('/homes/bdoc3/my_data/spmel_data/vocalSet_subset_unnormed/train/', file_name[:3], os.path.basename(file_name)))
speech_rtvc_arr = np.load(librispeech_file_path)

plot_numpy(npss_arr, title='NPSS (40 feats)', rot_num=3)
plot_numpy(rtvc_arr, title='RTVC (40 feats)', rot_num=3)
plot_numpy(autovc_arr, title='AutoVC (80 feats)', rot_num=3)
plot_numpy(speech_rtvc_arr, title='RTVC speech(40 feats)', rot_num=3)
```
